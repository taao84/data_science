# 

* 1. Run a cross validation in your dataset following the rule of:
    * 1.1. First take 10% train - 90% Test
        Run n iterations on your data with this arrangement.
        
        My accuracy metric will go from 0 to 100% (0 to 1).
        
        So in a 2D chart I could paint my accuracy values for train and test in this 10% Train

```python
x_train = [10]
y_train = [0.9]

x_test = [10]
y_test = [0.2]

plt.plot(x_train, y_train, 'o-')
plt.plot(x_test, y_test, 'o-')

plt.xticks(np.arange(0, 100, 10))
plt.xlim((0, 100))
plt.ylim((0, 1))

plt.title('Accuracy metrics')
plt.xlabel('Train dataset % usage')
plt.ylabel('Accuracy values')
plt.grid(True)
```

* 2. Keep running until you have the entire dataset in your train data:
    * 2.1. First take 90% train - 10% Test
        
        So in a 2D chart I could paint my accuracy values for train and test for all my values of Train size data.

After this you will one the following escenarios:

** 1. Train data accuracy is high but Test accuracy is still Low **

This means that my model is overfitting the Train dataset but is not at generalizing. 
The important thing here is to observe that the Test dataset got to a point where my improvement was flat

```python
x_train = list(range(10,100, 10))
y_train = [0.9, 0.8, 0.79, 0.78, 0.75, 0.74, 0.74, 0.74, 0.74]

x_test = list(range(10,100, 10))
y_test = [0.1, 0.15, 0.16, 0.20, 0.22, 0.35, 0.35, 0.35, 0.35]

plt.plot(x_train, y_train, 'o-')
plt.plot(x_test, y_test, 'o-')

plt.xticks(np.arange(0, 100, 10))
plt.xlim((0, 100))
plt.ylim((0, 1))

plt.title('Accuracy metrics')
plt.xlabel('Train dataset % usage')
plt.ylabel('Accuracy values')
plt.grid(True)
```


** 2. Train data accuracy is far from the 100% but gets close to the Test accuracy **

This means that my model has a bad prediction rate. 

```python
x_train = list(range(10,100, 10))
y_train = [0.9, 0.85, 0.80, 0.75, 0.75, 0.755, 0.750, 0.748, 0.746]

x_test = list(range(10,100, 10))
y_test = [0.9, 0.85, 0.80, 0.75, 0.75, 0.755, 0.750, 0.748, 0.746]

plt.plot(x_train, y_train, 'o-')
plt.plot(x_test, y_test, 'o-')

plt.xticks(np.arange(0, 100, 10))
plt.xlim((0, 100))
plt.ylim((0, 1))

plt.title('Accuracy metrics')
plt.xlabel('Train dataset % usage')
plt.ylabel('Accuracy values')
plt.grid(True)
```


** 3. Train accuracy is high but Test accuracy is low, but test accuracy does not flatten (Keeps increasing) **

This means that my data is not enough. I could try to improve my model, but if I have tried several models already then most likely I need more data.


** 4. Train accuracy is high and Test accuracy is high **

This is the ideal scenario where I have a good accuracy in my training and the model generalizes well in the Test data.

**NOTE:** One thing to keep in mind is that you need to be careful that you have not leaked the Test data and now you have learned it. Which means you will not generalize well with new data.
